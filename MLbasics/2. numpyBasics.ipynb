```python
import numpy as np
np_array=np.array([1,2,3,4,5])
print(np_array)
print(type(np_array))

b=np.array([[1,2,3],[4,5,6]])
print(b)
print(type(b))

```

```python
x=np.zeros((4,5))
print(x)
y=np.full((3,3),7) # full with 7
print(y)
z=np.eye(6)
print(z) #identity matrix

ab=np.random.random((3,5)) #all between 0 and 1
print(ab)

aa=np.random.randint(10,100,(2,8))#random integers
print(aa)
```

```python
#array of evenly spaced values
d=np.linspace(0,10,5)
print(d)

e=np.arange(0,10,3) #specify the steps
print(e)
```

```python
#numpy list
list=[1,2,3,4,5,6,7]
np_array=np.asarray(list)
print(np_array)
type(np_array)

#analyze a numpy array
c=np.random.randint(10,90,(5,5))
print(c)
print(c.shape)
print(c.ndim)
print(c.size)
print(c.dtype)
```

```python
#numpy operations

a=np.random.randint(5,25,(3,3))
b=np.random.randint(5,25,(3,3))
print(a)
print(b)
print(a+b)
print(np.add(a,b))
```
