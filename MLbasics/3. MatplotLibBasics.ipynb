# Wave

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x)')
plt.show()
```

parabolla

```python
x=np.linspace(-10,10,10)
y=x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('x^2 parabola')
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'gx')
plt.show()
```

multiple plot in single space

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r+')
plt.show
```

# Barplot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Latin','Bangla']
people=[40,23,15,67]
ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('people')
plt.title('barplot')
plt.show()
```

Pie chart

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Latin','Bangla']
people=[40,23,15,67]
ax.pie(people,labels=languages,autopct= '%1.1f%%')
plt.show()
```

Scatterplot

```python
x=np.linspace(0,10,50)
y=np.sin(x)
z=np.cos(x)
plt.scatter(x,y)
plt.scatter(x,z,color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('scatterplot')
plt.show()
```

3D scatter plot

```python
temp=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
